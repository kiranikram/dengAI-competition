```python
import pandas as pd
import numpy as np


# data viz
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sns

# apply some cool styling
plt.style.use("ggplot")
rcParams["figure.figsize"] = (12, 6)
```

```python
train_features = pd.read_csv("/Users/mussayababbasshamsi/Desktop/Kiran/dengAI-competition/data2/dengue_features_train.csv",index_col=[0, 1, 2])
```

```python

def split_data_by_city(df):
    sj = df.loc['sj']
    iq = df.loc['iq']

    return sj , iq
```

```python
sj_features , iq_features = split_data_by_city(train_features)
```

```python
num_cols = sj_features.select_dtypes(include=np.number).columns.tolist()


corrmat = sj_features.corr()
hm = sns.heatmap(
    corrmat,
    cbar=True,
    annot=True,
    square=True,
    fmt=".2f",
    annot_kws={"size": 10},
    yticklabels=sj_features.columns,
    xticklabels=sj_features.columns,
    cmap="Spectral_r",
)
plt.show()
```

```python
sj_features = sj_features[['reanalysis_relative_humidity_percent','station_avg_temp_c','station_min_temp_c','reanalysis_specific_humidity_g_per_kg','reanalysis_dew_point_temp_k']]
```

```python
num_cols = sj_features.select_dtypes(include=np.number).columns.tolist()
```

```python

for col in num_cols:
    print(col)
    print("Skew :", round(sj_features[col].skew(), 2))
    plt.figure(figsize=(15, 4))
    plt.subplot(1, 2, 1)
    sj_features[col].hist(grid=False)
    plt.ylabel("count")
    plt.subplot(1, 2, 2)
    sns.boxplot(x=sj_features[col])
    plt.show()
```

```python
num_cols_iq = iq_features.select_dtypes(include=np.number).columns.tolist()


corrmat = iq_features.corr()
hm = sns.heatmap(
    corrmat,
    cbar=True,
    annot=True,
    square=True,
    fmt=".2f",
    annot_kws={"size": 10},
    yticklabels=sj_features.columns,
    xticklabels=sj_features.columns,
    cmap="Spectral_r",
)
plt.show()
```
